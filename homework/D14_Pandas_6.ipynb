```python
#Col --> Index: .stack()
#Index --> Col: .unstack()

import pandas as pd
import numpy as np
data = pd.DataFrame(np.arange(6).reshape((2,3)),index=pd.Index(['Ohio','Colorado'],name='state')
                   ,columns=pd.Index(['one','two','three'],name='number'))

data
```

```python
result = data.stack() #將資料列旋轉為行
result
```

```python
result.unstack() #inverse operation, 回復原狀
```

```python
#欄位名稱 --> 欄位值：.melt(id_vars=不需要被轉換的欄位名稱, value_vars = 需要被轉換的欄位名稱)

df1 = pd.DataFrame({'Name':{0:'John', 1:'Bob', 2:'Shiela'}, 
                   'Course':{0:'Masters', 1:'Graduate', 2:'Graduate'}, 
                   'Age':{0:27, 1:23, 2:21}}) 

df1
```

```python
df1.melt(id_vars = 'Name')
```
