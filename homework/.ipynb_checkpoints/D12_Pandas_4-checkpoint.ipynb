```python
#折線圖

import numpy as np
import pandas as pd

#np.random.randn(200) # ~Z(0,1)
#pd.date_range('1/1/2020', periods = 200) #產生時間序列 2020-01-01 開始共200天

ts = pd.Series(np.random.randn(200), index = pd.date_range('1/1/2020', periods = 200))
#產生一個series, 每一天對應一個值, 共200天

ts.cumsum().plot() #累進加總
```

```python
#資料多個欄位, specify columns in DataFrame

df = pd.DataFrame(np.random.randn(200, 3), index = pd.date_range('1/1/2020', periods = 200), columns = ['A', 'B','C'])
#df: 200 x 3

df.cumsum().plot()
```

```python
#Piechart

A = pd.Series(np.random.rand(4), index = ["a", "b", "c", "d"], name = "title")
#產生一個Series, 有4個值編號a~d

A.plot.pie()
```

```python
#barchart

B = pd.DataFrame(np.random.rand(10,2), columns = ["a", "b"])
B.plot.bar()
```

```python
#boxplot

B.plot.box()
```

```python
B.boxplot()
```

```python
#scatter plot
B.plot.scatter(x = 'a', y = 'b')
```

題目 : 將資料夾中 boston.csv 讀進來，並用圖表分析欄位。
    
1.畫出箱型圖，並判斷哪個欄位的中位數在 300~400 之間？

2.畫出散佈圖 x='NOX', y='DIS' ，並說明這兩欄位有什麼關係？

```python
#Q1

import pandas as pd

boston = pd.read_csv('Boston.csv')

boston.boxplot(figsize = (10,8)) #看出TAX, B 中位數在300~400
```

```python
#Q2
import matplotlib.pyplot as plt
plt.scatter(boston.NOX, boston.DIS)
plt.xlabel('NOX')
plt.ylabel('DIS')

#NOX 越大 DIS越小
```
