```python
# plot

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 180) #0~179
y = np.sin(x * np.pi / 180.0)


plt.xlim(-30,390)
plt.ylim(-1.5,1.5)

plt.xlabel("x-axis") 
plt.ylabel("y-axis") 
plt.title("The Title") 

plt.plot(x, y)
plt.show()
```

```python
#scatter plot

X = np.random.normal(0, 1, 100) #mu=0, sigma=1
Y = np.random.normal(0, 1, 100)

plt.scatter(X, Y, marker = '+', s = 100, alpha = 0.5, c = 'r')
#s: size of point (20), c:color, 
#alpha: 透明度(0~1), marker
```

```python
# bar plots

x = np.arange(0, 10, 0.7) #0~9.8 (每隔0.7)
y = np.arange(0, 10, 0.7)

plt.bar(x, y)
```

```python
#Q1

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 180)
y_cos = np.cos(x)

plt.plot(x, y_cos)

plt.savefig("cos.jpg", dpi = 300, format = "jpg")
```
